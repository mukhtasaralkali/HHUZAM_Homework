Create a House class that represents a house; attributes:


*   area
*   number of rooms
*   number of bathrooms
*   number of floors
*   has garage
*   has garden
*   any other attributes that you think is important



instance methods:

*   price calculation

There should be at least 1 default parameter

(4 points)

```python
class House:
    def __init__(self, area=100, rooms=3, bathrooms=2, floors=1, garage=False, garden=False, location="Obuda District III"):
        self.area = area
        self.rooms = rooms
        self.bathrooms = bathrooms
        self.floors = floors
        self.garage = garage
        self.garden = garden
        self.location = location

    #the method to calculate price
    def calculate_price(self):
        price = self.area * 1000
        if self.garage:
            price += 50000000
        if self.garden:
            price += 30000
        if self.location == "Obuda District X":
            price += 10000
        return price


#Creating an object
house1 = House(150, 4, 3, 2, True, True, "Obuda District III")

print("House Location:", house1.location)
print("Area:", house1.area, "m²")
print("Rooms:", house1.rooms)
print("Estimated Price:", house1.calculate_price(),"HUF")
```

Create functions (decide whether it should be an instance method of one of the classes, or should be just defined in the program) to answer


*   whether the person can buy the house,
*   whether the person's family can live in that house (enough room),
*   whether the house has enough space (garages) for the cars in the family

(5 points)


```python
class House:
    def __init__(self, area=100, rooms=3, bathrooms=2,
                 floors=1, garage=False, garden=False,
                 location="Kerulet X"):

        self.area = area
        self.rooms = rooms
        self.bathrooms = bathrooms
        self.floors = floors
        self.garage = garage
        self.garden = garden
        self.location = location

    def calculate_price(self):
        price = self.area * 1000
        vat = price * 0.27
        return price + vat


class Person:
    def __init__(self, name, age, married,
                 children, cars, money):

        self.name = name
        self.age = age
        self.married = married
        self.children = children
        self.cars = cars
        self.money = money

        if married:
            self.family_members = children + 2
        else:
            self.family_members = children + 1


def can_buy_house(person, house):               #Functions
    return person.money >= house.calculate_price()

def enough_rooms(person, house):
    return house.rooms >= person.family_members

def enough_garage(person, house):
    if house.garage and person.cars <= 1:
        return True
    elif not house.garage and person.cars > 0:
        return False
    return True


house1 = House(150,4,3,2,True,True,"Kerulet X") #Instances
person1 = Person("Balaz",35,True,2,1,300000)


print("Can buy house:", can_buy_house(person1, house1))
print("Enough rooms:", enough_rooms(person1, house1))
print("Enough garage:", enough_garage(person1, house1))
```

```python
class Animal:
    def __init__(self, name, age, color):
        self.name = name
        self.age = age
        self.color = color


    def eat(self):                     #animal's attitude
        print(self.name, "is eating.")
    def sleep(self):
        print(self.name, "is sleeping.")
    def make_sound(self):
        print("makes a sound.")

class Dog(Animal):
    def __init__(self, name, age, color, breed):
        super().__init__(name, age, color)
        self.breed = breed
    def make_sound(self):
        print(self.name, "says: Woof woof")
    def guard_house(self):
        print(self.name, "is guarding the house.")

class Cat(Animal):
    def __init__(self, name, age, color, indoor):
        super().__init__(name, age, color)
        self.indoor = indoor

    def make_sound(self):
        print(self.name, "says: Meow")

    def climb(self):
        print(self.name, "climbs wall.")

class Bird(Animal):
    def __init__(self, name, age, color, wing_span):
        super().__init__(name, age, color)
        self.wing_span = wing_span

    def make_sound(self):
        print(self.name, "says: Kurkkkk!")

    def fly(self):
        print(self.name, "is flying.")

dog1 = Dog("Bingo", 5, "Brown", "german shepherd") #Object
cat1 = Cat("Maya", 3, "White", True) #Object
bird1 = Bird("Macaw", 2, "Blue", 35) #Object

dog1.eat()
dog1.make_sound()
dog1.guard_house()

print()

cat1.sleep()
cat1.make_sound()
cat1.climb()

print()

bird1.eat()
bird1.make_sound()
bird1.fly()

#All of these animals shares common attributes in the main class which includes; name, age and color, which are common in these animals.
#"Eat() and Sleep()" method is necessary, because all the animals eat and sleep
#"Make_sound()" each of the animals has a different sound, so the method is overriden in each sub-class.
```

###B) (6 points)

Create classes and use inheritance to calculate the area of different shapes (specifically, circle and square, you can add triangle or anything else if you want).

Requirements:


1.	Create a base class called Shape that has a method named area().

This method should not have any implementation
2. Create at least two subclasses:

Circle — which takes a radius as a parameter

Square — which takes a side as a parameter

write the area-calculating function for both (or for all)

3. Create objects of both Circle and Square, and print their areas.

Extra: Store these objects in a list and use a loop to print the area of each shape.

```python
import math

#Base class
class Shape:
    def area(self):
        pass

class Circle(Shape): #The subclass 1
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return math.pi * self.radius ** 2


class Square(Shape): #Subclass 2 taking side of the parameter
    def __init__(self, side):
        self.side = side
    def area(self):
        return self.side * self.side


class Triangle(Shape): #Extra shape
    def __init__(self, base, height):
        self.base = base
        self.height = height
    def area(self):
        return 0.5 * self.base * self.height

circle1 = Circle(5)
square1 = Square(4)
triangle1 = Triangle(6,8)

print("Circle area =", circle1.area())
print("Square area =", square1.area())
print("Triangle area =", triangle1.area())

print()

shapes = [circle1, square1, triangle1] #Using list and loop for storing
for shape in shapes:
    print("Area:", shape.area())
```

Expected output: ['a', 'e', 'e', 'u', 'a', 'i', 'o']

Combine two lists into tuples

```python
names = ["Anna", "Bob", "Charlie"]
scores = [85, 92, 78]
```

Expected output: [('Anna', 85), ('Bob', 92), ('Charlie', 78)]

```python
names = ["Anna", "Bob", "Charlie"]
scores = [85, 92, 78]

combined = []

for i in range(len(names)):
    combined.append((names[i], scores[i]))

print(combined)
```

Count character frequency

```python
text = "banana"
```

Expected output: {'b': 1, 'a': 3, 'n': 2}

```python
text = "banana"

count = {}

for letter in text:
    if letter in count:
        count[letter] += 1
    else:
        count[letter] = 1

print(count)
```
